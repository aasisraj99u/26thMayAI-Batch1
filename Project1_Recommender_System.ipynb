### Importing Necessary Modules

```python
import numpy as np
import warnings
import pandas as pd
import tkinter as tk
import time
import matplotlib.pyplot as plt
%matplotlib inline
warnings.filterwarnings('ignore')

cols = ['userId','item_id','rating','timestamp']  # u.data dataset column names
df = pd.read_csv('u.data',sep='\t',names=cols)   # reading u.data file


movie_titles = pd.read_csv('u.item',sep='|',encoding='latin-1',names=['item_id','movie title']+[str(i) for i in range(0,22)]) # reading u.item file
```

### Creating movie wise mean rating table and Adding freq of each movie 

```python
rating= pd.DataFrame(data.groupby(by='movie title')['rating'].mean().sort_values(ascending=False))
rating['Count'] = data['movie title'].value_counts()
rating.sort_values(by='Count',ascending=False)
```

### Pivot table between userId and movie title

```python
pivot_df = data.pivot_table(index='userId',columns='movie title',values='rating')
pivot_df
```

### GUI of Movie Recommender System

```python
app= tk.Tk()
app.title('Recommender System')
app.geometry('600x400')
app.config(bg='white')


f= tk.Frame(app,width=560,height=350,bd=5,bg='#FFFFE0',relief="ridge").place(x=20,y=25)
tk.Label(f,text="Watched Movie's : ",font=("quiche sans", 12),bg='#FFFFE0',fg='#ff9933').place(x=80,y=120)
tk.Label(f,text="Recommendations : ",font=("quiche sans", 12),bg='#FFFFE0',fg='#ff9933').place(x=80,y=220)
tk.Label(f,text="Also watch out : ",font=("quiche sans", 12),bg='#FFFFE0',fg='#ff9933').place(x=80,y=260)

title =tk.Label(f, text="Movie Recommendation System", font=("quiche sans", 15, "bold"), bg="#324ca8",
                      fg="white").place(x=50, y=35, width=500, height=35)
movie_var = tk.Variable(app)
tk.Entry(f,textvariable= movie_var,width=40).place(x=240,y=120)

recomm_var = tk.Variable(app)

tk.Label(app,textvariable=recomm_var).place(x=240,y=222)
also_try_var = tk.Variable(app)

tk.Label(f,textvariable=also_try_var).place(x=240,y=262)

def find_recomm():
        movie = movie_var.get().lower().rstrip()
        if movie:
            try:
                movie = movie_titles['movie title'][movie_titles['movie title'].apply(lambda x: movie in x.lower())].values[0]
                movie_var.set(movie)
            except IndexError:
                movie_var.set('Movie not found')
                recomm_var.set('No recommendation')
                also_try_var.set('No recommendation')
            
            else:
                corr_df = pd.DataFrame(pivot_df.corrwith(pivot_df[movie]), columns=['Correlation'])
                corr_df.dropna(inplace=True)
                corr_df = corr_df.join(rating['Count'])
                corr_df['mul_corr_count'] = corr_df['Correlation'] * corr_df['Count']
                recomm_var.set(corr_df[corr_df['Count']>200].sort_values(by='Correlation',ascending=False).index[1])
                also_try_var.set(corr_df[corr_df['Count']>200].sort_values(by='Correlation',ascending=False).index[2])
        else:
            movie_var.set('Please enter something')
            recomm_var.set('No recommendation')
            also_try_var.set('No recommendation')
def Clear():
    movie_var.set('')
    recomm_var.set('Type your movie above..')
    also_try_var.set('Type your movie above..')
    
tk.Button(app,text='Find Suggestions',command = find_recomm,bg='#33cc33').place(x=180,y=170)
tk.Button(app,text='Clear',width=10,command = Clear,bg='#33cc33').place(x=300,y=170)
recomm_var.set('Type your movie above..')
also_try_var.set('Type your movie above..')
app.mainloop()
```
