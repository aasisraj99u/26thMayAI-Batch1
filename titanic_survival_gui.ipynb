```python
import tkinter as tk
import pandas as pd
```

```python
app= tk.Tk()
app.title('Titanic Survival Prediction System')
app.geometry('700x500')
app.config(bg='white')


f= tk.Frame(app,width=660,height=450,bd=5,bg='#FFFFE0',relief="ridge").place(x=20,y=25)
title =tk.Label(f, text="Titanic Survival Prediction System", font=("quiche sans", 15, "bold"), bg="#ff9933",
                      fg="white").place(x=90, y=35, width=500, height=35)

tk.Label(f,text="Pclass : ",font=("quiche sans", 12),bg='#FFFFE0',fg='#ff9933').place(x=80,y=110)
tk.Label(f,text="Gender : ",font=("quiche sans", 12),bg='#FFFFE0',fg='#ff9933').place(x=80,y=150)
tk.Label(f,text="Age : ",font=("quiche sans", 12),bg='#FFFFE0',fg='#ff9933').place(x=80,y=195)
tk.Label(f,text="Total Parents/Children together : ",font=("quiche sans", 12),bg='#FFFFE0',fg='#ff9933').place(x=80,y=295)
tk.Label(f,text="Embarked Point : ",font=("quiche sans", 12),bg='#FFFFE0',fg='#ff9933').place(x=80,y=245)
tk.Label(f,text="Survival Prediction : ",font=("quiche sans", 12),bg='#FFFFE0',fg='#ff9933').place(x=80,y=410)


Age= tk.IntVar(app)
Parch = tk.IntVar(app)
survived_pred = tk.IntVar(app)


gender_type = ['male','female']
gender_selected = tk.StringVar(app)

Pclass_type = [1,2,3]
pclass_selected = tk.IntVar(app)

embarked_type = ['C','Q','S']
embarked_selected = tk.StringVar(app)

tk.OptionMenu(app,pclass_selected,*Pclass_type).place(x=240,y=110)
tk.OptionMenu(app,gender_selected, *gender_type).place(x=240,y=150)
tk.Entry(f,textvariable= Age,width=40,borderwidth=3).place(x=240,y=195)
tk.Entry(f,textvariable= Parch,width=40,borderwidth=3).place(x=340,y=295)
tk.OptionMenu(app,embarked_selected, *embarked_type).place(x=240,y=245)

tk.Label(app,textvariable=survived_pred,font=("quiche sans", 12)).place(x=240,y=410)

def find_recomm():
    if(gender_selected.get()=='male'):
        gender = 1
    else:
        gender = 0
    
    C=0
    Q=0
    S=0
    if(embarked_selected.get()=='C'):
        C = 1
    elif(embarked_selected.get()=='Q'):
        Q = 1
    else:
        S=1
        
    query = pd.DataFrame({
        'Pclass' : [pclass_selected.get()],
        'Age' : [Age.get()],
        'male': [gender],
        'Parch': [Parch.get()],
        'C': [C],
        'Q': [Q],
        'S': [S],
    })
    
    model = pd.read_pickle('TitanicSurvivalPrediction')
    prediction = int(model.predict(query)[0])
    if(prediction == 1):
        survived_pred.set('Survived')
    else:
        survived_pred.set('Died')

def Clear():
    pclass_selected.set(0)
    Age.set('')
    gender_selected.set('')
    Parch.set('')
    embarked_selected.set('')
    survived_pred.set(0)
    
tk.Button(app,text='Predict',command = find_recomm,bg='#33cc33',width=10).place(x=220,y=350)
tk.Button(app,text='Clear',width=12,command = Clear,bg='#33cc33').place(x=340,y=350)

app.mainloop()
```
