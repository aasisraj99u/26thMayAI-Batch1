```python
from tkinter import *
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, 
NavigationToolbar2Tk)
# Create object
root = Tk()
df = pd.read_csv('treadmil-users.csv')
# Adjust size
root.geometry()
  
# Change the label text
def show():
    
    fig = Figure(figsize = (5, 5),
                 dpi = 100)
    plot1 = fig.add_subplot(111)
    
    
    # plotting the graph
    if(column3.get()=='Histogram'):
        df[column1.get()].value_counts().plot(kind='hist', legend=True, ax=plot1)
        plt.title("Histogram of {}".format(column1.get()))
    elif(column3.get() == 'Barplot'):
        df[column1.get()].value_counts().plot(kind='bar', legend=True, ax=plot1)
        
    elif(column3.get() == 'Boxplot'):
        plot1.boxplot(df[column1.get()])
        
    elif(column3.get() == 'Scatter'):
        fig, ax = plt.subplots()
        ax.scatter(df[column1.get()],df[column2.get()])
        plt.xlabel(column1.get() + ' -->')
        plt.ylabel(column2.get() + ' -->')
        plt.title("Scatter plot b/w {} and {}".format(column1.get(),column2.get()))
    else:
        df[column1.get()].value_counts().plot.pie(autopct='%.1f%%',figsize=(8,8),ax=plot1) 
        plt.title("PieChart of {}".format(column1.get()))
        plt.show()
    # creating the Tkinter canvas
    # containing the Matplotlib figure
    canvas = FigureCanvasTkAgg(fig,
                               master = root)  
    canvas.draw()
  
    # placing the canvas on the Tkinter window
    canvas.get_tk_widget().pack()
  
    # creating the Matplotlib toolbar
    toolbar = NavigationToolbar2Tk(canvas,
                                   root)
    toolbar.update()
  
    # placing the toolbar on the Tkinter window
    canvas.get_tk_widget().pack()


options = [
    "Product",
    "Age",
    "Gender",
    "Education",
    "MaritalStatus",
    "Usage",
    "Fitness",
    "Income",
    "Miles"
]

options1 = [
    "Histogram",
    "Barplot",
    "Scatter",
    "Boxplot",
    "PieChart",
]

  
# datatype of menu text
column1 = StringVar()
column2 = StringVar() 
column3 = StringVar() 
# initial menu text
column1.set( "Product" )
column2.set("Product")
column3.set("Histogram")
# Create Dropdown menu
drop = OptionMenu( root , column1 , *options )
drop.place(x=10,y=15)

drop1 = OptionMenu( root , column2 , *options )
drop1.place(x=110,y=15)

drop2 = OptionMenu( root , column3 , *options1 )
drop2.place(x=50,y=80)


# Create button, it will change label text
button = Button( root , text = "PLOT" , command = show ).place(x=80,y=150)
root.mainloop()
```
